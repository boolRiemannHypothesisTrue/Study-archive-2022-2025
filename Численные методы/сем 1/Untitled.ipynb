```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
x=np.array([-0.7, -0.31, 0.08, 0.47, 0.86, 1.25, 1.64, 2.03, 2.42, 2.81, 3.2])
y=np.array([4.166, -2.278, -3.172, -0.506, 2.748, 2.665, 1.353, -0.294, -1.613, -2.223, 4.04])
```

```python
def genumsys(m,x, y): 
    C=np.zeros((m+1, m+1))
    b=np.zeros(m+1)
    for k in range (m+1): 
        b[k]=np.sum(y*x**k)
        for j in range (m +1):
            C[k,j] = np.sum(x**(k+j))
    return C, b
```

```python
def P(x , a):
    r = 0
    for i in range (len(a)):
        r+=a[i]*x**i
    return r
```

```python
def sigma(x , y, a):
    s = 0
    for i in range(len(x)):
        s+=(P(x[i], a) - y[i])**2
    s/=len(x)
    s =np.sqrt(s)
    return s
```

```python
a = []
s = []
for i in range (0,5):
    C , b = genumsys(i,x,y)
    a.append(np.linalg.solve(C,b))
    s.append(sigma(x,y,a[i]))
```

```python
print( a , s)
```
