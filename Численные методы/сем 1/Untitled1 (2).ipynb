```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
def f(x):
    return 5*np.sin(x**2)
```

```python
def l(x,xft,i):
    P = 1
    for j in range (len(xft)):
        if i != j :
            P*=(x-xft[j])/(xft[i]-xft[j])
    return P
```

```python
def L(x , xft, yft):
    s = 0
    for i in range (len(xft)):
        s += yft[i]*l(x,xft,i)
    return s
```

```python
n = 10
xft = np.linspace(0,4,n)
yft = f(xft)
print(L(3,xft,yft))
x1=np.linspace(0,4,1000)
plt.plot(x1,f(x1))
plt.plot(x1,L(x1,xft,yft))
```

```python
from scipy.interpolate import CubicSpline
spline = CubicSpline(xft,yft)
plt.plot(x1,spline(x1))
plt.plot(x1,f(x1))
plt.show
```

```python
plt.plot(x1,abs(f(x1)-L(x1,xft,yft)))
plt.show
```

```python
plt.plot(x1,abs(f(x1)-spline(x1)))
plt.show
```
